# Cleaning casenumber and date columns:

```python
def transformar(x):
    try:
        return x.replace('.', '-')
    except:
        return np.nan
```

```python
sharks.casenumber = sharks.casenumber.apply(transformar)

sharks.head()
```

```python
# Cleaning 'casenumber' column:

def filter_date(x):

    value = str(x)
    pattern = r"\d{4}\-\d{2}\-\d{2}"

    match = re.search(pattern, value)
    if match:
        clean_date = match.group()
        return clean_date
    else:
        return np.nan
    
```

```python
sharks.casenumber = sharks.casenumber.apply(filter_date)

sharks.head()
```

```python
def filter_date2(fecha):
    fecha=str(fecha)
    try:
        year, month, day = fecha.split('-')
        if month == '00' or day == '00' or len(year) != 4:
            return np.nan
        #pd.to_datetime(fecha)
        return fecha
    except:
        return np.nan
```

```python
sharks.casenumber = sharks.casenumber.apply(filter_date2)

sharks.head()
```

```python
sharks.casenumber.isnull().sum()
```

```python
sharks = sharks.dropna(subset=['casenumber'])

sharks.tail()
```

There are too many unique values in the 'species' column. To fix this we are going to create a function to filter the white sharks

```python
sharks.species.nunique()
```

```python
sharks.species.unique()
```

```python
def clean_white(sharks,species):
    
    has_white = sharks.species.str.contains('white', case =False)
    
    sharks.loc[has_white, species] = 'white shark'
    
    return sharks
```

```python
sharks = clean_white(sharks, 'species')
```

```python
def clean_hammer(sharks,species):
    
    has_hammer = sharks.species.str.contains('hammer', case =False)
    
    sharks.loc[has_hammer, species] = 'hammer shark'
    
    return sharks

sharks = clean_hammer(sharks,'species')

sharks.species.unique()
```

```python
def clean_tiger(sharks,species):
    
    has_tiger = sharks.species.str.contains('tiger', case =False)
    
    sharks.loc[has_tiger, species] = 'tiger shark'
    
    return sharks

sharks = clean_tiger(sharks,'species')

sharks.species.unique()
```

```python
def clean_bull(sharks,species):
    
    has_bull = sharks.species.str.contains('bull', case =False)
    
    sharks.loc[has_bull, species] = 'bull shark'
    
    return sharks

sharks = clean_bull(sharks,'species')

sharks.species.unique()
```

```python
sharks = sharks[sharks.species!='']

sharks.species.nunique()
```

```python
sharks.species.value_counts(normalize=True) * 100
```

### As our OBJECTIVE COLUMN is fatal(y/n) we are going to delete wherever it is unknow data 

```python
sharks['fatal(y/n)'].value_counts(normalize=True) * 100
```

```python
def delete_incorrect(sharks, column, allowed_values):
    
    sharks = sharks[sharks['fatal(y/n)'].isin(allowed_values)]
    
    return sharks

allowed_values = ['n', 'y']

sharks = delete_incorrect(sharks,'fatal(y/n)',allowed_values )

sharks.shape
```

```python
sharks['fatal(y/n)'].value_counts(normalize=True) * 100
```

```python
def change_fatal_col (sharks, column):
    
    sharks[column]=sharks[column].map({'y' : 'Yes', 'n' : 'No'})
    
    return sharks

sharks = change_fatal_col(sharks,'fatal(y/n)')

sharks['fatal(y/n)'].value_counts(normalize=True) * 100
```
